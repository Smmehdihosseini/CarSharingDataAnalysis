```python
def get_duration(city, Unix_startTime, Unix_endTime):
    
    query = [
        {
            '$match' : {
                'city': city,
                'init_time': {'$gte': Unix_startTime, '$lte': Unix_endTime}
            }
        },
        {
            '$project':{
                '_id':0,
                'city':1,
                'duration':{
                    '$ceil': {
                        '$subtract': ['$final_time', '$init_time']}
                }
            }
        },
        {
            '$group': {
                '_id': '$duration'}
        }
    ]
    
    return query

def hist_book_n_park(city_dict):
    fig, axes = plt.subplots(1, 1, figsize=(6, 6))
    axes.hist(city_dict['bookings'],
                bins=4000,
                density=True,
                histtype='step',
                cumulative=True,
                label='Bookings',
                color='black',
                linewidth=2,
                fill=True,
                alpha=0.3)
    axes.set_xscale('log')
    axes.hist(city_dict['parkings'],
                bins=4000,
                density=True,
                histtype='step',
                cumulative=True,
                label='Parkings',
                color='orange',
                linewidth=2,
                fill=True,
                alpha=0.5)
    
    axes.set_xscale('log')
    axes.legend(loc='upper left')
    axes.set_title('CDF of Parkings and Bookings In '+ str(city_dict['city']), fontsize=12)
    axes.set_xlabel('Duration [Sec]', fontsize=10)
    axes.set_ylabel('CDF', fontsize=10)
    axes.grid(True, which='both')
    plt.tight_layout()
    plt.savefig('Task1_CDF_' + str(city_dict['city']) + '.png', dpi=600)
    plt.show()
    
cities = ['Seattle', 'Milano Enjoy', 'Milano']
         
for city in cities:
    city_dict = {'city': city, 'bookings':[], 'parkings':[]}
    
    if 'Enjoy' in city:
        startingDate = dt(2017, 12, 1, 0)
        endingDate = dt(2018, 2, 28, 0)
        Unix_startTime = dt.timestamp(startingDate)
        Unix_endTime = dt.timestamp(endingDate)
        query = get_duration('Milano', Unix_startTime, Unix_endTime)
        i_booking = enj_booking_perm
        i_parking = enj_parking_perm
        
    else:
        startingDate = dt(2017, 12, 1, 0)
        endingDate = dt(2018, 2, 28, 0)
        Unix_startTime = dt.timestamp(startingDate)
        Unix_endTime = dt.timestamp(endingDate)
        query = get_duration(city, Unix_startTime, Unix_endTime)
        i_booking = c2g_booking_perm
        i_parking = c2g_parking_perm

    bookings_duration = i_booking.aggregate(query)
    city_dict['bookings'] =[duration['_id'] for duration in list(bookings_duration)]

    
    parking_duration=i_parking.aggregate(query)
    city_dict['parkings']=[duration['_id'] for duration in list(parking_duration)]
    hist_book_n_park(city_dict)
```

```python
def get_duration_WeekDay(city, day):
    query = [
                {
                    '$match' : {'city': city, 'init_time': {'$gte': Unix_startTime, '$lte':
                    Unix_endTime}}
                },
                {
                    '$project':{
                                '_id':0,
                                'city':1,
                                'duration':{'$ceil': {'$subtract': ['$final_time', '$init_time']}},
                                'WeekDay': { '$dayOfWeek': "$init_date" },
                                }
                },
                {
                    '$match': {'WeekDay': day}
                },
                {
                    '$group': { '_id': '$duration'}
                },
                {
                    '$sort':{'_id':1}
                }]
    return query

city = 'Milano'

plt.figure(figsize=(12,5))

for day in range(1, 8):
    bookings_durations = c2g_booking_perm.aggregate(get_duration_WeekDay(city, day))
    bookings = [duration['_id'] for duration in list(bookings_durations)]
    plt.hist(bookings,
             bins=4000,        
             density=True,
             histtype='step',
             cumulative=True,
             linewidth=2,
             alpha=1)
    plt.xscale('log')
        
plt.legend(["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"], loc='lower right')
plt.title('CDF of Booking Duration Per Each Day of the Week in ' + city, fontsize=12)
plt.xlabel('Duration [Sec]', fontsize=10)
plt.ylabel('CDF', fontsize=10)
plt.grid(True, which='both')
plt.tight_layout()
plt.savefig('Task1_CDF_PERDAY'+city+'_book.png', dpi=600)
plt.show()

plt.figure(figsize=(12, 5))

for day in range(1, 8):
    bookings_durations = c2g_parking_perm.aggregate(get_duration_WeekDay(city, day))
    bookings = [duration['_id'] for duration in list(bookings_durations)]
    plt.hist(bookings,
             bins=4000,        
             density=True,
             histtype='step',
             cumulative=True,
             linewidth=2,
             alpha=1)
    plt.xscale('log')
        
plt.legend(["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"], loc='lower right')
plt.title('CDF of Parking Duration Per Each Day of the Week in ' + city, fontsize=12)
plt.xlabel('Duration [Sec]', fontsize=10)
plt.ylabel('CDF', fontsize=10)
plt.grid(True, which='both')
plt.tight_layout()
plt.savefig('Task1_CDF_Weekly_'+city+'_park.png', dpi=600)
plt.show()
```

### S2_T2

```python
def plot_daily_book_park(city_dict, filtered):

    
    if filtered==True:
        filter_value = 'Filtered'
    else:
        filter_value = ''
        
    fig, axs = plt.subplots(1, 1, figsize=(12, 5))
    date_ticks = np.arange("2017-12-01","2018-02-28", dtype ='datetime64')

    booking_date, no_bookings = zip(*sorted(city_dict['bookings'].items()))
    parking_date, no_parkings = zip(*sorted(city_dict['parkings'].items()))

    axs.plot(booking_date, no_bookings, label='Bookings', color='green', linewidth=0.5)
    axs.plot (parking_date, no_parkings, label='Parkings', color='red', linewidth=0.5)
    axs.legend(loc='upper left')
    axs.set_xticks(date_ticks)
    axs.set_xticklabels(labels=date_ticks, rotation=90, ha='right')
    axs.set_title(filter_value + ' Number of Booked / Parked Cars Per Hour in ' + str(city_dict['city']), fontsize=12)
    axs.set_xlabel('Days', fontsize=10)
    axs.set_ylabel('No of Booked / Parked Cares ', fontsize=10)
    axs.grid(True, which='both')
    plt.tight_layout()
    plt.xticks(rotation=90)
    plt.savefig('Task2_CarsPerHour_' + city_dict['city'] + '_' + filter_value + '.png', dpi=600)
    plt.show()
    
def get_cars_per_Hour(city, Unix_startTime, Unix_endTime):
    
    query = [
        {
        '$match': { 'city': city,
                    'init_time': {'$gte': Unix_startTime, '$lte': Unix_endTime}
                  }
        },
        {
        '$project' : { '_id':0,
                       'city': 1,
                       'hourOfDay': {'$dateFromParts': {'year':{'$year':'$init_date'}, 'month': {'$month':'$init_date'}, 'day': {
                       '$dayOfMonth': '$init_date'}, 'hour': {'$hour': '$init_date'}}},
                       'duration':{'$ceil': {'$subtract': ['$final_time',
                       '$init_time']} },
                       'moved': {'$ne': [
                    {'$arrayElemAt': ['$origin_destination.coordinates', 0]},
                    {'$arrayElemAt': ['$origin_destination.coordinates', 1]}]
                    }
                    }
        },
        {
        '$group': { '_id': '$hourOfDay',
                    'count':{'$sum':1}
        }
            
        },
        {
        '$sort': {'_id': 1}
        }
    ]
    
    return query

cities=['Seattle', 'Milano Enjoy', 'Milano']

for city in cities:
    city_dict = {'city': city, 'bookings':{}, 'parkings':{}}
    if 'Enjoy' in city:
        startingDate = dt(2017, 12, 1, 0)
        endingDate = dt(2018, 2, 28, 0)
        Unix_startTime = dt.timestamp(startingDate)
        Unix_endTime = dt.timestamp(endingDate)
        query = get_cars_per_Hour('Milano', Unix_startTime, Unix_endTime)
        i_booking = enj_booking_perm
        i_parking = enj_parking_perm
        
    else:
        startingDate = dt(2017, 12, 1, 0)
        endingDate = dt(2018, 2, 28, 0)
        Unix_startTime = dt.timestamp(startingDate)
        Unix_endTime = dt.timestamp(endingDate)
        query = get_cars_per_Hour(city, Unix_startTime, Unix_endTime)
        i_booking = c2g_booking_perm
        i_parking = c2g_parking_perm

    daily_bookings = i_booking.aggregate(query)
    for item in daily_bookings:
        city_dict['bookings'][item['_id']] = item['count']
        
    daily_parkings = i_parking.aggregate(query)
    for item in daily_parkings:
        city_dict['parkings'][item['_id']] = item['count']
        
    plot_daily_book_park(city_dict, False)
```

### S2_T3 & T4

```python
def get_cars_per_Hour_bookFilter(city, Unix_startTime, Unix_endTime):
    
    query = [
        {
            '$match': { 'city': city,
                    'init_time': {'$gte': Unix_startTime, '$lte': Unix_endTime}
                  }
        },
        {
            '$project' : { '_id':0,
                       'city': 1,
                       'hourOfDay': {'$dateFromParts': {'year':{'$year':'$init_date'}, 'month': {'$month':'$init_date'}, 'day': {
                       '$dayOfMonth': '$init_date'}, 'hour': {'$hour': '$init_date'}}},
                       'duration':{'$ceil': {'$subtract': ['$final_time',
                       '$init_time']} },
                       'moved': {'$ne': [
                    {'$arrayElemAt': ['$origin_destination.coordinates', 0]},
                    {'$arrayElemAt': ['$origin_destination.coordinates', 1]}]
                    }
                    }
        },
        {
            '$match': {
                    'moved': True,
                    'duration': {'$gte': 4*60, '$lte': 2*60*60}
                    }
        },
        
        {
            '$group': { '_id': '$hourOfDay',
                    'count':{'$sum':1}
        }
            
        },
        {
            '$sort': {'_id': 1}
        }
    ]
    return query


def get_cars_per_Hour_parkFilter(city, Unix_startTime, Unix_endTime):
    
    query = [
        {
            '$match': { 'city': city,
                        'init_time': {'$gte': Unix_startTime, '$lte': Unix_endTime}
                  }
        },
        {
            '$project' : { '_id':0,
                       'city': 1,
                       'hourOfDay': {'$dateFromParts': {'year':{'$year':'$init_date'}, 'month': {'$month':'$init_date'}, 'day': {
                       '$dayOfMonth': '$init_date'}, 'hour': {'$hour': '$init_date'}}},
                       'duration':{'$ceil': {'$subtract': ['$final_time',
                       '$init_time']} },
                       'moved': {'$ne': [
                    {'$arrayElemAt': ['$origin_destination.coordinates', 0]},
                    {'$arrayElemAt': ['$origin_destination.coordinates', 1]}]
                    }
                    }
        },
        {
            '$match': {
                    'duration': {'$gte': 4*60}
                    }
        },
        
        {
            '$group': { '_id': '$hourOfDay',
                    'count':{'$sum':1}
        }
            
        },
        {
            '$sort': {'_id': 1}
        }
    ]
    return query 

cities=['Seattle', 'Milano Enjoy', 'Milano']

for city in cities:
    city_dict = {'city': city, 'bookings':{}, 'parkings':{}}
    if 'Enjoy' in city:
        startingDate = dt(2017, 12, 1, 0)
        endingDate = dt(2018, 2, 28, 0)
        Unix_startTime = dt.timestamp(startingDate)
        Unix_endTime = dt.timestamp(endingDate)
        book_pipe = get_cars_per_Hour_bookFilter('Milano', Unix_startTime, Unix_endTime)
        park_pipe = get_cars_per_Hour_parkFilter('Milano', Unix_startTime, Unix_endTime)
        i_booking = enj_booking_perm
        i_parking = enj_parking_perm
        
    else:
        startDate = dt(2017, 12, 1, 0)
        endDate = dt(2018, 2, 28, 0)
        startUnixTime = dt.timestamp(startDate)
        endUnixTime = dt.timestamp(endDate)
        book_pipe = get_cars_per_Hour_bookFilter(city, Unix_startTime, Unix_endTime)
        park_pipe = get_cars_per_Hour_parkFilter(city, Unix_startTime, Unix_endTime)
        i_booking = c2g_booking_perm
        i_parking = c2g_parking_perm
    
    daily_bookings = i_booking.aggregate(book_pipe)
    for item in daily_bookings:
        city_dict['bookings'][item['_id']] = item['count']
        
    daily_parkings = i_parking.aggregate(park_pipe)
    for item in daily_parkings:
        city_dict['parkings'][item['_id']] = item['count']
        
    plot_daily_book_park(city_dict, True)
```
